```python
array = [3, -1, -7, -4, -2, 10, -6]
def find_negative_elements_between_min_max(arr):
    max_index = arr.index(max(arr)) 
    min_index = arr.index(min(arr))
    start = min(max_index, min_index) + 1  
    end = max(max_index, min_index)
    result = [x for x in arr[start:end] if x < 0]
    return result
result = find_negative_elements_between_min_max(array)
print(":", result)
```

```python
array = [5, 3, 4, 8, 2, 7, 1]
def sort_even_index_elements(arr):
    even_index_elements = [arr[i] for i in range(0, len(arr), 2)]
    even_index_elements.sort()
    result = arr[:]
    even_index_pointer = 0
    for i in range(0, len(arr), 2):
        result[i] = even_index_elements[even_index_pointer]
        even_index_pointer += 1
    return result
result=sort_even_index_elements(array)
print(":", result)
```
